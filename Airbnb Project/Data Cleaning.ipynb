```python
# histogram of price values
prices = listings_df['price']
plt.hist(prices, bins=30, color='skyblue', edgecolor='black')
plt.grid(True, linestyle='--', alpha=0.1)
plt.xlabel('price')
plt.ylabel('Frequency')

plt.show()
```

```python
# detect outliers in prices

Q1_price = listings_df['price'].quantile(0.25)
Q3_price = listings_df["price"].quantile(0.75)
IQR_price = Q3_price - Q1_price

price_outliers = listings_df[(listings_df["price"] < (Q1_price - 1.5 * IQR_price)) | (listings_df["price"] > (Q3_price + 1.5 * IQR_price))]

# Q1 is 76
# Q3 is 200
# no downwards outliers, Q1-1.5*IQR goes into negative prices , we therefore only have outliers beyond Q3, or more precisely beyong Q3+1,5*IQR = 386 

price_outliers = price_outliers.sort_values(by="price", ascending=True)

print(price_outliers[['listing_id', 'price']])

# smallest outliers start at 387$ price
# 4581 outliers
```
